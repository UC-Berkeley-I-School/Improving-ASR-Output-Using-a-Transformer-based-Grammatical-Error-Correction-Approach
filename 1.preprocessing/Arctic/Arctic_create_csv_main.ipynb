# combine all csvs to one main csv

```python
main = pd.concat([bdl, clb, rms, slt], axis=0)

main.head()
```

```python
main['actor'].value_counts()
```

```python
assert len(main[main['actor'] == 'bdl']) == len(bdl)
assert len(main[main['actor'] == 'clb']) == len(clb)
assert len(main[main['actor'] == 'rms']) == len(rms)
assert len(main[main['actor'] == 'slt']) == len(slt)
```
