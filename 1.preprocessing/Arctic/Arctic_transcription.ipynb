# Read the transcription file

```python
df = pd.read_csv('/content/drive/MyDrive/266/Data/Raw_Data/Arctic/cmuarctic.txt', sep='"', header=None, names=['filename', 'transcription', 'na'])

df.head()
```

Clean up the df

```python
# remove the last column that's not needed
df.drop(columns=['na'], inplace=True)

df.head()
```

```python
# cleanup the filename column
df['filename'] = df['filename'].str.split('(').str[1]

df.head()
```

```python
# take a look at the filename
df['filename'].value_counts()
```

```python
# to confirm the first half of the filename column are the same for all instances
first_half = df['filename'].str.split('_').str[0]

first_half.value_counts()
```

```python
# keep only the second half of the filename
df['filename'] = df['filename'].str.split('_').str[1]
df['filename'] = df['filename'].str.strip()

df.head()
```

```python
# confirm all filenames are unique
assert len(df) == df['filename'].nunique()
```
